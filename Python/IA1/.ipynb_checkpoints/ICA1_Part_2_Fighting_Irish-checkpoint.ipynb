Write a program that prints the numbers from 1 to 100. But for **multiples of three print “Fighting”** instead of the number and for the **multiples of five print “Irish”**. For numbers which are multiples of **both three and five print “Fighting Irish”**. 

### Hint:
* You can use `for` loop to iterate between numbers
* You will need to use `%` operator

```python
for i in range(1,101):
    if i%3 == 0 and i%5 == 0:
        print("Fighting Irish")
    elif i%3 == 0:
        print("Fighting")
    elif i%5 == 0:
        print("Irish")
    else:
        print(i)
```
