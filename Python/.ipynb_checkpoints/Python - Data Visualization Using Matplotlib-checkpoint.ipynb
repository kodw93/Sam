### Dot Plot

#### 1. Load the dataset mpg.csv, and create the following dot plot using matplotlib. This dotplot looks at the average miles per gallon for each car manufacturer.

```python
data = pd.read_csv("mpg(1)(1).csv")
data.head()
```

```python
# preparing the data
# create a dataframe that shows the average of city mpg for each car manufacturer

datavis = data.groupby('manufacturer')['cty'].mean()

datavis = datavis.reset_index()
datavis.sort_values('cty',inplace = True)
datavis
```

```python
# creating the visualization

plt.figure()
plt.scatter(x = datavis['cty'], 
            y = datavis['manufacturer'].str.title(), 
            color = 'firebrick', alpha = 0.6)
ax = plt.gca()
ax.hlines(y = datavis.manufacturer.str.title(), 
          xmin = 10, xmax = 26,
          alpha = 0.4, 
          linestyle = '-.',
          linewidth = 0.8)
plt.title ("Average Car City Mileage")
plt.xlabel("Mileage")
sns.despine(left = True, bottom = True)  # gets rid of the box
ax.yaxis.set_ticks_position('none')  # this gets rid of the ticks
plt.show()
```

### Diverging Bars

#### 2. Create the following diverging bar plot using matplotlib. This plot looks at the average miles per gallon (z-scores) for each car manufacturer.

```python
# create a new column "cty_z" that holds the standardized scores for the cty variable
data['cty_z'] = (data.cty - data.cty.mean())/data.cty.std()   
# when creating new column, have to usd df["new column"]
data.head()
```

### all of these lead to same result of accessing the column
##### data.cty
##### data['cty']
##### data.loc[:,'cty']

### if want to access more than 1 column,
##### data[['cty','hwy']]
##### data.loc[:,['cty','hwy']]

```python
datavis = data.groupby('manufacturer')['cty_z'].\
                mean().\
                reset_index().\
                sort_values('cty_z')
datavis
# \ just means the code continues in the next line
```

```python
# create the color variable based on the sign of the cty_z variable
datavis['col']=["green" if x>0 else "red" for x in datavis.cty_z]
datavis
```

```python
plt.figure()
ax = plt.gca()   #get current axis
ax.hlines(y = datavis.manufacturer.str.title(),
          xmin = 0, 
          xmax = datavis.cty_z,
     
          color = datavis.col, alpha = 0.4,
          linewidth = 2)
ax.hlines(y = datavis.manufacturer.str.title(),
          xmin = -2, 
          xmax = 2,
          color = "grey",
          alpha = 0.4,
          linewidth = 0.8,
          linestyle = ":")
plt.title("Average Car Mileage in the City")
plt.show()
```

### Pairwise Plots

#### 4. Create a pairwise plot to show the relationship between the following numerical variables in the mpg.csv dataset.

```python
datavis = data[['displ','cty','hwy']]
datavis.head()
```

```python
datavis.columns = ["Displacement","City Mileage","Highway Mileage"]
```

```python
plt.figure(figsize=(5,5))
sns.pairplot(datavis)
plt.show()
```

#### 5. Update the previous pairwise plot to the regression lines.

```python
plt.figure(figsize=(5,5))
sns.pairplot(datavis,
             kind = "reg")

plt.show()
```

#### 6. Update the initial pairwise plot to show the relationships among the different car engine sizes (cyl).

```python
datavis = data[['displ','cty','hwy','cyl']]
datavis.columns = ["Displacement","City Mileage","Highway Mileage","Cylinder"]
plt.figure(figsize=(5,5))
sns.pairplot(datavis,
             hue = "Cylinder",
            x_vars= ["Displacement","City Mileage","Highway Mileage"],
            y_vars= ["Displacement","City Mileage","Highway Mileage"])

plt.show()
```
