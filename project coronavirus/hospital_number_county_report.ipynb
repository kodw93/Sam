### As the data is directly from the federal government medicare website, I was comfortable with the credibility of the source.

```python
# source: https://data.medicare.gov/Hospital-Compare/Hospital-General-Information/xubh-q36u

import pandas as pd
from sodapy import Socrata

BASE_URL = "data.medicare.gov"
PATH = 'xubh-q36u'

us_state_abbrev = {
    'Alabama': 'AL',
    'Alaska': 'AK',
    'American Samoa': 'AS',
    'Arizona': 'AZ',
    'Arkansas': 'AR',
    'California': 'CA',
    'Colorado': 'CO',
    'Connecticut': 'CT',
    'Delaware': 'DE',
    'District of Columbia': 'DC',
    'Florida': 'FL',
    'Georgia': 'GA',
    'Guam': 'GU',
    'Hawaii': 'HI',
    'Idaho': 'ID',
    'Illinois': 'IL',
    'Indiana': 'IN',
    'Iowa': 'IA',
    'Kansas': 'KS',
    'Kentucky': 'KY',
    'Louisiana': 'LA',
    'Maine': 'ME',
    'Maryland': 'MD',
    'Massachusetts': 'MA',
    'Michigan': 'MI',
    'Minnesota': 'MN',
    'Mississippi': 'MS',
    'Missouri': 'MO',
    'Montana': 'MT',
    'Nebraska': 'NE',
    'Nevada': 'NV',
    'New Hampshire': 'NH',
    'New Jersey': 'NJ',
    'New Mexico': 'NM',
    'New York': 'NY',
    'North Carolina': 'NC',
    'North Dakota': 'ND',
    'Northern Mariana Islands': 'MP',
    'Ohio': 'OH',
    'Oklahoma': 'OK',
    'Oregon': 'OR',
    'Pennsylvania': 'PA',
    'Puerto Rico': 'PR',
    'Rhode Island': 'RI',
    'South Carolina': 'SC',
    'South Dakota': 'SD',
    'Tennessee': 'TN',
    'Texas': 'TX',
    'Utah': 'UT',
    'Vermont': 'VT',
    'Virgin Islands': 'VI',
    'Virginia': 'VA',
    'Washington': 'WA',
    'West Virginia': 'WV',
    'Wisconsin': 'WI',
    'Wyoming': 'WY'
}
abbrev_us_state = dict(map(reversed, us_state_abbrev.items()))


client = Socrata(BASE_URL, None)
results = client.get(PATH, limit=10000)
hospital = pd.DataFrame.from_records(results)
```

### As seen below, there are no null values assoicated with the hospital names at the counties.

```python
hospital.isnull().sum()
```

**No null values for the columns we need**

```python
hospital = hospital[['hospital_name', 'state', 'county_name' ]]
hospital.isnull().sum()
```

**1569 unique counties found in the dataset**

```python
hospital['county_name'].nunique()
```

```python
hospital_count =  hospital.groupby(['state', 'county_name']).size().rename('count').reset_index()
hospital_count
```

#### The belows lines of code are to format the data so that it corresponds to the exisiting data in MongoDB. 

```python
hospital_count['state']= hospital_count['state'].map(abbrev_us_state)
hospital_count
```

```python
hospital_count['county_name'] =  hospital_count['county_name'].str.lower().str.title()
hospital_count
```

```python
hospital_count['county_name'] = hospital_count['county_name'].astype(str) + ' County'
hospital_count
```
