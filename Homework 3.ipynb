```python
# Importing packages needed for the homework
import pandas as pd
import matplotlib.pyplot as plt
```

#### 2. Create a line chart that shows President’s Trump approval rating as follows. The approval line is green, the disapproval line is orange, and the title is bold of fontsize 14. In addition, the plot’s figure size is 11x6

```python
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()

# this was advised by the notebook to register
# if I don't have these, there appears a warning sign, but the code runs 
```

```python
plt.figure (figsize = (11,6))
plt.plot(trump["Date"], trump["Approve"],
        color = "green",
        label = "Approval")
plt.plot(trump["Date"], trump["Disapprove"],
        color = "orange",
        label = "Disapproval")
plt.title ("President Trump's Approval Rating",
          fontsize = 14, fontweight = "bold")
plt.legend(loc = "upper left")
plt.ylabel ("President's Rating (%)")
plt.show()
```

#### 3. Add a horizontal line at “2018-01-01”. The vertical line should be dashed and grey.

```python
plt.figure(figsize=(11,6))
plt.plot(trump["Date"],trump["Approve"],
        color = "green",
        label = "Approval")
plt.plot(trump["Date"],trump["Disapprove"],
        color = "orange",
        label = "Disapproval")
plt.legend(loc = "upper left")
plt.title("President Trump's Approval Rating", 
         fontsize = 14,
         fontweight = "bold")
plt.axvline(x= "2018-01-01", 
            color = "grey", 
            linestyle = "--")
plt.ylabel ("President's Rating (%)")
plt.show()
```

#### 5. Add a horizontal line at the 50% level approval as follows. The line should be dotted red.

```python
plt.figure (figsize = (11,6))
plt.plot("Date","Approve",
         data = trump,
         color = "green",
         label = "Approval")
plt.plot("Date","Disapprove",
         data = trump,
         color = "orange",
         label = "Disapproval")
plt.legend(loc="upper left")
plt.title("President Trump's Approval Rating", 
         fontsize = 14,
         fontweight = "bold")
plt.ylim(30,70)
plt.axvline(x= "2018-01-01", 
            color = "grey", 
            linestyle = "--")
plt.ylabel ("President's Rating (%)")
plt.axhline(y= 50, color = "red", linestyle = "dotted")
plt.show()
```

#### 6. Use the matplotlib method axvspan() to add a vertical span to show the government shutdown period between Dec 22, 2018 and Jan 1, 2019. The span should be grey and the alpha level should be 0.3.

```python
plt.figure (figsize = (11,6))
plt.plot("Date","Approve",
         data = trump,
         color = "green",
         label = "Approval")
plt.plot("Date","Disapprove",
         data = trump,
         color = "orange",
         label = "Disapproval")
plt.legend(loc="upper left")
plt.title("President Trump's Approval Rating", 
         fontsize = 14,
         fontweight = "bold")
plt.ylabel ("President's Rating (%)")
plt.ylim(30,70)
plt.axvline ( x = "2018-1-1", color = "grey", linestyle = "dashed")
plt.axhline(y= 50, color = "red", linestyle = "dotted")
plt.axvspan("2018-12-22", "2019-1-25", facecolor = "grey", alpha = 0.3)
plt.show()
```

#### 7. Add the text labels on the plot as follows. The font size for all text is 12

```python
plt.figure (figsize = (11,6))
plt.plot("Date","Approve",
         data = trump,
         color = "green",
         label = "Approval")
plt.plot("Date","Disapprove",
         data = trump,
         color = "orange",
         label = "Disapproval")
plt.legend(loc="upper left")
plt.title("President Trump's Approval Rating", 
         fontsize = 14,
         fontweight = "bold")
plt.ylabel ("President's Rating (%)")
plt.ylim(30,70)
plt.axvline ( x = "2018-1-1", color = "grey", linestyle = "dashed")
plt.axhline(y= 50, color = "red", linestyle = "dotted")
plt.axvspan("2018-12-22", "2019-1-25", facecolor = "grey", alpha = 0.3)
plt.annotate("Dec. 22 - Jan. 25 \nGovernment shutdown", ("2018-12-10", 65), size = 12)
plt.annotate("One Year \nin office", ("2018-1-1",65), size = 12)
plt.annotate("57% \nDisapprove", ("2019-8-1",57), color = "orange", size = 12)
plt.annotate("39% \nApprove", ("2019-8-1",39), color = "green", size = 12)
plt.show()
```
